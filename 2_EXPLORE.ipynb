```python
import pandas as pd

# Carga cada archivo y revisa
datasets = {
    "wellbore": "data/raw/wellbore.csv",
    "fields_overview": "data/raw/fields_overview.csv",
    "production_monthly": "data/raw/production_monthly.csv",
    "reserves": "data/raw/reserves.csv",
    "investment_yearly": "data/raw/investment_yearly.csv",
    "investment_expected": "data/raw/investment_expected.csv",
}

for name, path in datasets.items():
    df = pd.read_csv(path, low_memory=False)
    print(f"\n{'='*40}")
    print(f"{name.upper()}")
    print(f"Filas: {len(df)} | Columnas: {len(df.columns)}")
    print(f"Nulls:\n{df.isnull().sum()[df.isnull().sum() > 0]}")
    print(f"Tipos:\n{df.dtypes}")
```

```python
import pandas as pd

prod = pd.read_csv("data/clean/production_monthly.csv")

# Produccion anual total
annual = prod.groupby("year")["oil_mill_sm3"].sum().reset_index()
print("Annual oil production (last 5 years):")
print(annual.tail(5))

# Meses disponibles por año
months = prod.groupby("year")["month"].nunique()
print("\nMonths available per year:")
print(months.tail(5))
```

```python
import pandas as pd

prod = pd.read_csv("data/clean/production_monthly.csv")

# Verificar años y meses disponibles
print("Años disponibles:", sorted(prod["year"].unique()))
print("\nMeses por año:")
print(prod.groupby("year")["month"].count())
```

```python
# Verificar totales mensuales agrupados
monthly_totals = prod.groupby(["year","month"]).agg(
    oil=("oil_mill_sm3","sum"),
    gas=("gas_mill_sm3","sum"),
    ngl=("ngl_mill_sm3","sum")
).reset_index()

print("2025 monthly totals:")
print(monthly_totals[monthly_totals["year"]==2025][["month","oil","gas","ngl"]].to_string())
```
